```python
import gdax
public_client = gdax.PublicClient()
products = public_client.get_products()
ids = [x['id'] for x in products]
times = {}
prices = {}
sizes = {}
for id in ids:
    times[id] = []
    prices[id] = []
    sizes[id] = []
```
