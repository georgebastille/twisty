```python
import gdax
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from collections import OrderedDict

from datetime import datetime, timedelta
from math import ceil
from time import sleep

from backtest import Trader, historic_prices
from algorithms import SMA, EMA, PPOAlgo, StochasticOscillator, MovingAverageAlgo, PriceWatchOscillator

%matplotlib inline
client = gdax.PublicClient()
```

```python
prices = historic_prices(client=client, product='BTC-EUR', num_candles=288, granularity=300)
ticks = len(prices)
print('{} prices'.format(ticks))

pd.Series(prices).plot(figsize=(20,10))
```

```python
look_back = 79
buy_hold = round((prices[-1]/prices[-look_back] * 100), 2)
print('Buy Hold return = {}%'.format(buy_hold))
```

```python
#PPO Backtest
potential = []
for slow_win in range(1, 30, 1):
    for fast_win in range(1, slow_win, 1):
        for signal_win in range(1, 10):
            
            ppoAlgo = PPOAlgo(slow_win=slow_win, 
                              fast_win=fast_win, 
                              signal_win=signal_win)
            
            trader = Trader(algo=ppoAlgo, verbose=False, trade_prob=1.0)
            final = 0
            trades = 0
            for price in prices[-look_back:]:
                final, trades = trader.tick(price)
            final = (round(final, 2))
            adjusted = final - trades/10.0
            potential.append((adjusted, final, trades, slow_win, fast_win, signal_win))
potential.sort(key=lambda tup: tup[0], reverse=True)
for sims in potential[:3]:
    print('{}, {}, {} ({}, {}, {})'.format(*sims))
```

```python
# Moving Average Backtest
potential = []
for slow in range (2, 50, 1):
    for fast in range(1, slow, 1):
        slow_avg = SMA(slow)
        fast_avg = SMA(fast)
            
        algo = MovingAverageAlgo(slow=slow_avg, 
                             fast=fast_avg,   
                             buy_cutoff=0.0,          
                             sell_cutoff=0.0)
            
        trader = Trader(algo=algo, verbose=False, trade_prob=1.0)
        final = 0
        trades = 0
        for price in prices[-look_back:]:
            final, trades = trader.tick(price)
        final = (round(final, 2))
        adjusted = final - trades/10.0
        potential.append((adjusted, final, trades, slow, fast))
potential.sort(key=lambda tup: tup[0], reverse=True)
for sims in potential[:3]:
    print('{}, {}, {} ({}, {})'.format(*sims))
```

```python
# Stochastic Oscillator Backtest
potential = []
for window in range (2, 30, 1):
    for smooth in range(2, 30, 1):
        algo = StochasticOscillator(window, smooth)
        trader = Trader(algo=algo, verbose=False, trade_prob=1.0)
        final = 0
        trades = 0
        for price in prices[-look_back:]:
            final, trades = trader.tick(price)
        final = (round(final, 2))
        adjusted = final - trades/10.0
        potential.append((adjusted, final, trades, window, smooth))
potential.sort(key=lambda tup: tup[0], reverse=True)
for sims in potential[:3]:
    print('{}, {}, {} ({}, {})'.format(*sims))
```

```python
# PriceWatch Oscillator Backtest
potential = []
for up_chg in np.arange(0.0, 0.05, 0.005):
    for down_chg in np.arange(0.0, 0.05, 0.005):
        for up_out_chg in np.arange(0.0, 0.05, 0.005):
            algo = PriceWatchOscillator(up_chg=up_chg, down_chg=down_chg, up_out_chg=up_out_chg)
            trader = Trader(algo=algo, verbose=False, trade_prob=0.8)
            final = 0
            trades = 0
            for price in prices[-look_back:]:
                final, trades = trader.tick(price)
            final = (round(final, 2))
            adjusted = final - trades/10.0
            potential.append((adjusted, final, trades, up_chg, down_chg, up_out_chg))
potential.sort(key=lambda tup: tup[0], reverse=True)
for sims in potential[:3]:
    print('{}, {}, {} ({}, {}, {})'.format(*sims))
```
