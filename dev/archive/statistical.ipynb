```python
import csv
with open('./data/GOOG-2.csv') as csvfile:
    spamreader = csv.reader(csvfile, delimiter=',')
    _ = spamreader.__next__() # Skip headers for now
    close = []
    for row in spamreader:
        close.append(row[4])
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
import numpy as np

def diff(vals, delta=-1):
    shift = np.roll(vals, delta)
    diff = np.subtract(shift, vals)
    return diff[:delta]

```

```python
close = np.array(close, dtype=np.float)
delta = diff(close)
```

```python
plt.plot(close)
plt.show()

plt.plot(delta[:50])
plt.show()

norm = np.correlate(delta, np.roll(delta, 0))
norm = 1
corr = []
for i in range(10):
    corr.append((np.correlate(delta, np.roll(delta, -i))/norm)[0])
plt.plot(np.array(corr, dtype=np.float))
plt.show()
```

```python
plt.hist(delta, bins=200)
plt.yscale('log')
plt.show()
```
